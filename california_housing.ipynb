```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def plot_metric_distribution(df, metric, target_range=30):
    # Set the style and color palette
    sns.set(style="whitegrid")
    palette = sns.color_palette("coolwarm", as_cmap=True)

    target = df[metric]

    # Plot the target with a cool color palette
    plt.figure(figsize=(10, 6))
    ax = sns.histplot(target, bins=target_range, kde=True, color=palette(0.5), legend=True)
    plt.title(f'Distribution of {metric.replace("_", " ").title()}', fontsize=16)
    plt.xlabel(metric.replace("_", " ").title(), fontsize=14)
    plt.ylabel('Frequency', fontsize=14)

    # calculate the increment for the intervals based on the target range
    increment = (target.max() - target.min()) / 30

    # Add median value intervals on the x-axis    
    median_intervals = np.arange(target.min(), target.max(), increment)
    labels = [f'[{interval:,.0f}, {interval + increment:,.0f})' for interval in median_intervals]
    print(labels)
    plt.xticks(median_intervals, rotation=45, labels=labels, ha='right')

    # Display the frequency on top of the bars
    for patch in ax.patches:
        height = patch.get_height()
        if height > 0:  # Only display text for visible bars
            ax.text(patch.get_x() + patch.get_width() / 2, height + 5, int(height),
                    ha='center', fontsize=10)

    plt.show()

# Example usage
plot_metric_distribution(df, 'median_house_value')
```

```python
features = ['median_income', 'total_rooms', 'housing_median_age', 'latitude']

for feature in features:
    plot_metric_distribution(df, feature, target_range = 20)

```

```python
import numpy as np

def plot_boxplot_with_mean_median(df, column):
    plt.figure(figsize=(12, 8))
    sns.set(style="whitegrid")

    # Boxplot
    sns.boxplot(x=df[column], color='skyblue')

    # Calculate mean and median
    mean_value = df[column].mean()
    median_value = df[column].median()

    # Add vertical lines for mean and median
    plt.axvline(mean_value, color='red', linestyle='--', label=f'Mean: {mean_value:.2f}')
    plt.axvline(median_value, color='green', linestyle='-', label=f'Median: {median_value:.2f}')

    plt.title(f'{column.replace("_", " ").title()} with Mean and Median', fontsize=18, color='darkblue')
    plt.xlabel(column.replace("_", " ").title(), fontsize=14, color='darkblue')
    plt.xticks(rotation=45, color='darkblue')
    plt.yticks(color='darkblue')
    plt.legend()
    plt.tight_layout()
    plt.show()

# Example usage
plot_boxplot_with_mean_median(df, 'median_house_value')
```

```python
features = ['median_income', 'total_rooms', 'housing_median_age', 'latitude']

for feature in features:
    plot_boxplot_with_mean_median(df, feature)
```

```python
from sklearn.preprocessing import StandardScaler # type: ignore
import torch.nn as nn
from sklearn.model_selection import train_test_split # type: ignore
import torch
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('data/california_housing_train.csv')

# Separate target and features
target = df['median_house_value']
features = df[['median_income', 'housing_median_age', 'latitude', 'total_rooms']]

# Normalize features and target
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

target_scaled = target / target.max()  # Scale target to be in the range [0, 1]

# Split the data
X_train, X_test, y_train, y_test = train_test_split(features_scaled, target_scaled, test_size=0.2, random_state=42)

device = torch.device("mps")

# Define the model
class LinearRegression(nn.Module):
    def __init__(self, input_size, output_size):
        super(LinearRegression, self).__init__()
        self.linear = nn.Linear(input_size, output_size, device=device)

    def forward(self, x):
        out = self.linear(x)  # linear layer
        return out

# Create the model
input_size = features.shape[1]
output_size = 1
model = LinearRegression(input_size, output_size)

# Define loss function and optimizer
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

# Define number of epochs
num_epochs = 2000

# Convert data to tensors
X_train = torch.tensor(X_train, dtype=torch.float32)
y_train = torch.tensor(y_train.values, dtype=torch.float32).unsqueeze(1)  # Add dimension for target
X_test = torch.tensor(X_test, dtype=torch.float32)
y_test = torch.tensor(y_test.values, dtype=torch.float32).unsqueeze(1)

# Train the model
for epoch in range(num_epochs):
    # Forward pass
    outputs = model(X_train)
    loss = criterion(outputs, y_train)

    # Backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}')

# Evaluate model
with torch.no_grad():
    predictions = model(X_test)
    test_loss = criterion(predictions, y_test)
    print(f'Test Loss: {test_loss.item():.4f}')

# Plot predictions
plt.figure(figsize=(12, 8))
plt.scatter(y_test, predictions.numpy())
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.title('True vs Predicted Values')
plt.show()



import datetime
# how can i save the model and load it later?
#could you change the name of the file to model.pth and include increment in the name of the file abd date?
torch.save(model.state_dict(), f'model_{epoch}-{datetime.datetime.now().strftime("%Y-%m-%d-%H-%M-%S")}.pth')
torch.save(model.state_dict(), f'model-latest.pth')

# model.load_state_dict(torch.load(f'model_{epoch}.pth'))
# how can i use the model to predict new values for the test dataset?


```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Load data
df = pd.read_csv('data/california_housing_train.csv')

# Separate target and features
target = df['median_house_value'].values
features = df[['median_income', 'housing_median_age', 'latitude', 'total_rooms']].values

# Normalize features and target
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
target_scaled = target / target.max()  # Scale target to be in the range [0, 1]

# Add intercept term
intercept = np.ones((features_scaled.shape[0], 1))
X = np.hstack((intercept, features_scaled))

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, target_scaled, test_size=0.2, random_state=42)

# Initialize coefficients
beta = np.zeros(X_train.shape[1])

# Define hyperparameters
learning_rate = 0.01
num_epochs = 2000

# Gradient descent
for epoch in range(num_epochs):
    # Compute predictions
    y_pred = X_train @ beta
    # Compute error
    error = y_pred - y_train
    # Compute gradient
    gradient = (1 / len(y_train)) * (X_train.T @ error)
    # Update coefficients
    beta -= learning_rate * gradient

    # Print loss every 100 epochs
    if (epoch + 1) % 100 == 0:
        mse = np.mean(error ** 2)
        print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {mse:.4f}')

# Evaluate on test set
y_test_pred = X_test @ beta
test_mse = np.mean((y_test - y_test_pred) ** 2)
print(f'Test MSE: {test_mse:.4f}')
```
