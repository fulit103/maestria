```python

import pandas as pd

# Cargar el dataset
file_path = 'desercion.xlsx'  # Asegúrate de que el archivo esté en el mismo directorio
data = pd.read_excel(file_path)

# Mostrar las primeras filas y la estructura del dataset
data.head(), data.info()
```

## Distribución de la Variable Objetivo `ESTADO_INSTITUCION`

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Graficar la distribución de la variable 'ESTADO'
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='ESTADO_INSTITUCION')
plt.title("Distribución de la Variable Objetivo: ESTADO")
plt.xlabel("Estado")
plt.ylabel("Frecuencia")
plt.xticks(rotation=45)
plt.show()
```

## Transformación de `ESTADO_INSTITUCION` en Variable Binaria `TARGET`

```python
# print unique values of ESTADO_INSTITUCION
print(data['ESTADO_INSTITUCION'].unique())
```

```python

# Convertir 'ESTADO' en una variable binaria para indicar deserción
data['TARGET'] = data['ESTADO_INSTITUCION'].apply(lambda x: 1 if '2. Permanece programa' in x else 0)

# Verificar la transformación
data['TARGET'].value_counts()
```

## Cálculo de Correlación con la Variable `TARGET`

```python

# Seleccionar sólo las columnas numéricas y la nueva columna 'TARGET'
data_numeric = data.select_dtypes(include=['int64', 'float64'])

# Calcular la matriz de correlación
correlation_matrix = data_numeric.corr()
correlation_with_target = correlation_matrix['TARGET'].sort_values(ascending=False)
correlation_with_target
```

## Identificación de Outliers

### Box plot for PUNTAJE_ICFES
### Box plot for EDAD

```python
import matplotlib.pyplot as plt

# Box plot for PUNTAJE_ICFES
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=data, y='PUNTAJE_ICFES')
plt.title('Box Plot of PUNTAJE_ICFES')

# Box plot for EDAD
plt.subplot(1, 2, 2)
sns.boxplot(data=data, y='EDAD')
plt.title('Box Plot of EDAD')

plt.tight_layout()
plt.show()

# Calculate IQR and identify outliers for PUNTAJE_ICFES
Q1_icfes = data['PUNTAJE_ICFES'].quantile(0.25)
Q3_icfes = data['PUNTAJE_ICFES'].quantile(0.75)
IQR_icfes = Q3_icfes - Q1_icfes
outliers_icfes = data[(data['PUNTAJE_ICFES'] < (Q1_icfes - 1.5 * IQR_icfes)) | (data['PUNTAJE_ICFES'] > (Q3_icfes + 1.5 * IQR_icfes))]

# Calculate IQR and identify outliers for EDAD
Q1_edad = data['EDAD'].quantile(0.25)
Q3_edad = data['EDAD'].quantile(0.75)
IQR_edad = Q3_edad - Q1_edad
outliers_edad = data[(data['EDAD'] < (Q1_edad - 1.5 * IQR_edad)) | (data['EDAD'] > (Q3_edad + 1.5 * IQR_edad))]

print("Outliers in PUNTAJE_ICFES:")
print(outliers_icfes)

print("\nOutliers in EDAD:")
print(outliers_edad)
```

```python
from sklearn.model_selection import train_test_split

# Define features and target
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
features = scaler.fit_transform(data[['PUNTAJE_ICFES', 'EDAD']])
target = data['TARGET']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Display the shapes of the resulting datasets
print("Training features shape:", X_train.shape)
print("Testing features shape:", X_test.shape)
print("Training target shape:", y_train.shape)
print("Testing target shape:", y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initialize the RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
rf_model.fit(X_train, y_train)

# Make predictions
y_pred = rf_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Print classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Print confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

# Análisis

- **Rendimiento**:

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Define features and target
features = data[['PUNTAJE_ICFES', 'EDAD']]
target = data['TARGET']

# Balancear las clases usando SMOTE
smote = SMOTE(random_state=42)
features_balanced, target_balanced = smote.fit_resample(features, target)

# Escalar las características
scaler = StandardScaler()
features_balanced_scaled = scaler.fit_transform(features_balanced)

# Dividir el dataset balanceado en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(features_balanced_scaled, target_balanced, test_size=0.2, random_state=42)

# Mostrar las formas de los datasets resultantes
print("Training features shape:", X_train.shape)
print("Testing features shape:", X_test.shape)
print("Training target shape:", y_train.shape)
print("Testing target shape:", y_test.shape)

# count the number of 0 and 1 in the target_balanced
print(target_balanced.value_counts())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initialize the RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
rf_model.fit(X_train, y_train)

# Make predictions
y_pred = rf_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Print classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Print confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

results = confusion_matrix(y_test, y_pred)

accuracy = results.trace() / results.sum()
precision = results[1, 1] / results[:, 1].sum()
recall = results[1, 1] / results[1, :].sum()
f1 = 2 * precision * recall / (precision + recall)

comparison_results = pd.DataFrame({
    'Modelo': ['Random Forest'],
    'Accuracy': [accuracy],
    'Precision': [precision],
    'Recall': [recall],
    'F1-Score': [f1],
    'Confusion Matrix': [results]
})

comparison_results.head()
```
