# Descomposicion matricial

Factorizacion de una matiz en productos de matices mas simples

## Descomposicion de valor singular (SVD)



```python
A = np.array([[3,1], [1,3]])

U, S, Vt = np.linalg.svd(A)

print("Matriz A:")
print(A)
print("Matriz U:")
print(U)
print("Matriz S:")
print(S)
print("Matriz Vt:")
print(Vt)


# get the initial matrix
A = U.dot(np.diag(S)).dot(Vt)

print("Matriz A:")
print(A)



```

# Derivadas y Gradientes

- Derivada: medida de la rapidez de cambio de una funcion en un punto
- Gradiente: vector de derivadas parciales de una funcion multivariable
- calculo en numpy

```python
import numpy as np
import matplotlib.pyplot as plt

# Función de dos variables
def f(x, y):    
    return x**2 + y**2

# Crear una malla de puntos
x = np.linspace(-5, 5, 20) # 20 puntos entre -5 y 5 
y = np.linspace(-5, 5, 20) # 20 puntos entre -5 y 5
X, Y = np.meshgrid(x, y) # Crear una malla de puntos
Z = f(X, Y) # Calcular el valor de la funcion en cada punto

# Calcular gradientes
df_dx = 2*X
df_dy = 2*Y

# Visualizar campo de gradientes

plt.figure()
plt.contour(X, Y, Z)
plt.quiver(X, Y, df_dx, df_dy)
plt.title('Campo de Gradientes')
plt.show() 

# could you plot the gradientes in 3d?
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function of two variables
def f(x, y):
    return x**2 + y**2

# Create a higher-resolution grid of points
x = np.linspace(-5, 5, 50)  # Increased from 20 to 50 points
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Calculate gradients
df_dx = 2 * X
df_dy = 2 * Y
df_dz = -1  # For surface z = f(x, y), the gradient in z is -1

# Calculate the magnitude of the gradient vectors
magnitude = np.sqrt(df_dx**2 + df_dy**2 + df_dz**2)

# Normalize the gradient vectors for visualization
df_dx_unit = df_dx / magnitude
df_dy_unit = df_dy / magnitude
df_dz_unit = df_dz / magnitude

# Create a 3D plot
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface with enhanced visualization
surf = ax.plot_surface(
    X, Y, Z,
    rstride=1, cstride=1,
    facecolors=plt.cm.coolwarm(Z / np.max(Z)),
    linewidth=0, antialiased=True, shade=True, alpha=0.8
)

# Add color bar for the surface
mappable = plt.cm.ScalarMappable(cmap=plt.cm.coolwarm)
mappable.set_array(Z)
fig.colorbar(mappable, shrink=0.5, aspect=10, pad=0.1, label='Function Value')

# Reduce the number of vectors to plot for clarity
skip = (slice(None, None, 5), slice(None, None, 5))

# Map magnitudes to colors for the gradient vectors
norm = plt.Normalize(magnitude.min(), magnitude.max())
colors = plt.cm.plasma(norm(magnitude[skip]))

# Plot the gradient vectors with colors based on magnitude
ax.quiver(
    X[skip], Y[skip], Z[skip],
    df_dx_unit[skip], df_dy_unit[skip], df_dz_unit[skip],
    length=0.5, normalize=False, colors=colors, linewidth=1.5, arrow_length_ratio=0.3
)

# Set labels and title with improved aesthetics
ax.set_xlabel('X', labelpad=15, fontsize=12)
ax.set_ylabel('Y', labelpad=15, fontsize=12)
ax.set_zlabel('Z', labelpad=15, fontsize=12)
ax.set_title('Gradient Vectors on 3D Surface', pad=20, fontsize=16)

# Adjust the viewing angle for a more dramatic effect
ax.view_init(elev=45, azim=135)

# Remove gridlines for a cleaner look
ax.grid(False)

# Set background color
ax.set_facecolor('white')
fig.patch.set_facecolor('white')

# Show the plot
plt.show()
```
