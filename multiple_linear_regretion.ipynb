```python
import numpy as np

def multiple_linear_regression(X, y, learning_rate=0.01, num_epochs=1000):
    # Add intercept term
    intercept = np.ones((X.shape[0], 1))
    X = np.hstack((intercept, X)) # X is n x (m+1) matrix where n is the number of samples and m is the number of features

    # Initialize coefficients
    beta = np.zeros(X.shape[1])

    for epoch in range(num_epochs):
        # Compute predictions
        y_pred = X @ beta
        # Compute error
        error = y_pred - y
        # Compute gradient
        gradient = X.T @ error / len(y)
        # Update coefficients
        beta = beta - learning_rate * gradient

    return beta

# Example usage
X = np.array([[1, 2], [3, 4], [5, 6]])
y = np.array([3, 7, 11])

beta = multiple_linear_regression(X, y)
print(beta)


#graph X and y 3d
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[:, 0], X[:, 1], y)

```
